```python
import pandas as pd
import matplotlib.pyplot as plt
from matplotlib import font_manager, rc
import seaborn as sns
import numpy as np
import random

font_path = 'c:/Windows/Fonts/malgun.ttf'
font_name = font_manager.FontProperties(fname=font_path).get_name()
rc('font', family=font_name)
```

```python
df = pd.read_csv('./analyze_Daegu_22.csv')
df = df[['매장명','인접_버스정류장수']]
df.set_index('매장명', inplace=True)
df
```

```python
line_num = int(df.mean())

fig = plt.figure(figsize=(10,5))
ax1 = fig.add_subplot(1,1,1)

sns.barplot(data=df, x='인접_버스정류장수', y=df.index, ax=ax1)
plt.title('대구 스타벅스 인접 버스정류장 수', fontsize=12, loc='left')
plt.legend(fontsize=8, ncol=4, loc='upper left')
plt.tick_params(width=0)
ax1.set_xlabel('인접 버스정류장 수', loc='right', fontsize=9)
plt.axvline(x=line_num, color='black', linewidth=1)
```
